#
creating input(x) and output(y) values from df

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

#
graphical visuation of each input feature v/s ouput('quality')

```python
sns.pairplot(data=df,y_vars='quality',x_vars=df.drop(columns='quality').columns)
plt.show()
```

#
spliting of training and testing data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

#
find coefficient and intercept

```python
print(list(zip(x,model.coef_)))

print('Intercept is',model.intercept_)
```

#
create dataframe of actual and predicted values of y

```python
df1=pd.DataFrame({"actual_value":y_test,"predicted_value":y_pred})
df1
```

#
calculating errors

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error
print("Mean Absolute error is",mean_absolute_error(y_test,y_pred))
print("Mean Absolute Percentage error is",mean_absolute_percentage_error(y_test,y_pred))
print("Mean Squared Error is",mean_squared_error(y_test,y_pred))

```
