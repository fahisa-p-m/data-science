> droping time column and columns with large missing values

```python
df.drop(['Time','Service_year_of_vehicle','Defect_of_vehicle','Work_of_casuality','Fitness_of_casuality'], axis = 1, inplace = True)
df.head()
```

> the missing values are all catogorical. creating an object for the catrgorical values

```python
categorical=[i for i in df.columns if df[i].dtype=='O']
print(categorical)
print(len(categorical))
```

> filling the columns with mode

```python
for i in categorical:
 df[i].fillna(df[i].mode()[0],inplace=True)
```

> encoding catogorical values using LabelEncoder





```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
for col in df.columns:
    if df[col].dtype==object:
        df[col]=le.fit_transform(df[col])
```

> splitting x and y values

```python
x=df.iloc[:,:-1]
x.head()
```

```python
y=df.iloc[:,-1]
y.head()
```

```python
y.value_counts()
```

> splitting into training data and testing data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

> creating model

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
knn=KNeighborsClassifier(n_neighbors=5)
nb=MultinomialNB()
svc=SVC()
tree=DecisionTreeClassifier(criterion='entropy')
forest=RandomForestClassifier(n_estimators=10, criterion="entropy")
lstmodel=[knn,nb,svc,tree,forest]
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
for i in lstmodel:
  print(i)
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print('***************************************************************')
  print(classification_report(y_test,y_pred))
  print('****************************************************************')
```
